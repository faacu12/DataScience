```python
import pandas as pd

def cargar_dataset(opcion):
    if opcion.upper() == "A":
        return pd.read_csv("marketing_campaign.csv", sep=";")
    elif opcion.upper() == "B":
        return pd.read_csv("ecommerce_customer_churn_dataset.csv", sep=";")
    else:
        return None

opcion = input("Ingrese A para el dataset de Marketing o B para el segundo dataset: ")

df = cargar_dataset(opcion)

if df is not None:
    print("Dataset cargado correctamente")
    display(df.head())
else:
    print("Opción inválida")

```
